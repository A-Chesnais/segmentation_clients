# Segmentation RFM

## Sélection des features

Dans un premier temps, on se concentrera sur les données 'order_count', 'days_since_last_order', 'order_value' pour effectuer une segmentation Récence, Fréquence et montant.

```python
data_RFM = data.loc[:,['order_count', 'days_since_last_order', 'order_value']].copy()
```

```python
RFM_pop_average = data_RFM.mean()
display(RFM_pop_average)
```

## Transformation log

On a également remarqué que la variable 'order_value' est très écrasée vers les petites valeurs, on effectuera un passage au log pour celle ci, ainsi que pour le nombre de commandes pour la même raison.

```python
data_RFM_log = data_RFM.copy()
data_RFM_log['order_value'] = np.log(data_RFM['order_value'] + 0.001)
data_RFM_log['order_count'] = np.log(data_RFM['order_count'] + 0.001)
```

```python
data_RFM_log.hist(figsize=(10,10), bins=50);
```

## Scaling des données

Les échelles étant très différentes pour les différentes variables, nous allons appliquer un StandardScaler sur les données. Cela pemettra d'obtenir la même moyenne et la même variance pour toutes les variables.

```python
scaler = StandardScaler().fit(data_RFM_log)
scaled_data_RFM = pd.DataFrame(scaler.transform(data_RFM_log), columns=data_RFM.columns, 
                               index=data_RFM.index)
```

```python
scaled_data_RFM.hist(figsize=(10,10), bins=50);
```

## Modélisation Kmeans

### Nombre de clusters optimal

Dans un premier temps on recherche un nombre de clusters optimal, en cherchant à minimiser la somme de la distance de chaque point d'un cluster à son centre (sum square error, sse). On choisit ainsi un nombre de clusters à partir duquel la réduction de la sse se modifie beaucoup moins fortement avec l'introduction d'un nouveau cluster. On testera différents nombres de clusters autour de cette valeur 'optimale'.

```python
sse = {}
for k in range(1, 11): 
    kmeans=KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data_RFM)
    sse[k] = kmeans.inertia_
```

```python
plt.title('Elbow criterion method chart')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
```

### Modèle à 3 clusters

Afin d'évaluer les clusters, on regardera les indicateurs suivants :  
- Les moyennes pour chaque variable au sein des différents clusters, relativement à celle de la population globale.
- La proportion de la population dans chaque segment
- La contribution de chaque segment au chiffre d'affaire
- Le score de silouhette

```python
kmeans3 = functions.kmeans_analysis(3, data_RFM, scaled_data_RFM, 'viridis')
```

Avec cette première segmentation en 3 parties on obtient les résultats suivants :
- Un premier segment (1), que l'on pourrait définir comme celui des gens qui ont récemment commandé et qui n'ont commandé qu'une seule fois
- Un seconde segment (0), qui correspond au clients qui ont commandé il y a longtemps et qui n'ont commandé qu'une seule fois
- Un segment (2), avec des clients qui commandent plusieurs fois et génèrent plus de valeurs que les autres.  
  
A noter que la dernière classe est très peu présente (3% des individus)

```python
print('kmeans: {}'.format(silhouette_score(data_RFM, kmeans3.labels_, metric='euclidean')))
```

Représentation spatiale des segments (scaled data) :

```python
fig = plt.figure();
ax = Axes3D(fig);
ax.scatter(scaled_data_RFM['days_since_last_order'], scaled_data_RFM['order_value'],
          scaled_data_RFM['order_count'], c=kmeans3.labels_);
ax.view_init(elev=10, azim=45)
ax.set_xlabel('days_since_last_order (scaled)', labelpad=15)
ax.set_ylabel('orders value (scaled)', labelpad=15)
ax.set_zlabel('orders count (scaled)')
```

### Modèle à 4 clusters

```python
kmeans4 = functions.kmeans_analysis(4, data_RFM, scaled_data_RFM, 'viridis')
```

Avec cette nouvelle segmentation en 4 parties, on obtient les résultats suivants :

- Un segment (0), qui correspond aux clients qui ont effectué une seule commande, mais d'un montant plus élevé que la moyenne.
- Un segment (1),  correspondant à des acheteurs à 'faible valeur' mais récents (commande unique avec une valeur inférieure à la moyenne)
- Un segment (2), où l'on retrouve les acheteurs "récidivistes", ayant acheté à de multiples reprises.
- Un segment (3) englobant les acheteurs à 'faible valeur', ayant commandé une seule fois, pour des montant inférieurs à la moyenne et ce il y a longtemps.  
 
A noter que le segment (0) correspond à 30% de la population mais contribue à plus de 50% du chiffre d'affaire.

```python
print('kmeans: {}'.format(silhouette_score(data_RFM, kmeans4.labels_, metric='euclidean')))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(scaled_data_RFM['days_since_last_order'], scaled_data_RFM['order_value'],
          scaled_data_RFM['order_count'], c=kmeans4.labels_)
ax.view_init(elev=10, azim=45)
ax.set_xlabel('days_since_last_order (scaled)', labelpad=15)
ax.set_ylabel('orders value (scaled)', labelpad=15)
ax.set_zlabel('orders count (scaled)')
```

### Modèle à 5 clusters

```python
kmeans5 = functions.kmeans_analysis(5, data_RFM, scaled_data_RFM, 'viridis')
```

En segmentant en 5 clusters, on retrouve de fortes similarités avec le modèle précédent en contenant 4. Seule 'nouveauté', le segment des clients n'ayant commandé qu'une seule fois récemment en maintenat séparé en deux : les montants de commande inférieur à la moyenne et ceux qui sont au niveau de celle ci.

```python
print('kmeans: {}'.format(silhouette_score(data_RFM, kmeans5.labels_, metric='euclidean')))
```

```python
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(scaled_data_RFM['days_since_last_order'], scaled_data_RFM['order_value'],
          scaled_data_RFM['order_count'], c=kmeans5.labels_)
ax.view_init(elev=10, azim=45)
ax.set_xlabel('days_since_last_order (scaled)', labelpad=15)
ax.set_ylabel('orders value (scaled)', labelpad=15)
ax.set_zlabel('orders count (scaled)')
```

A noter que le score de silouhette diminue avec l'augmentation du nombre de cluster, malgré une segmentation qui pourtant fait plus de sens. On se basera donc ensuite essentiellement sur l'interprétatibilité 'business' de la segmentation pour juger de la qualité de celle ci

## Définition finale des segments

### A partir des features utilisées :

```python
kmeans_RV_5 = functions.kmeans_analysis(5, data_RFM_RV, scaled_data_RFM_RV, 'viridis')
```

Si on reprend la description établie :
    
**Segment (0)** : les acheteurs **_'récidivistes'_**.  
Il s'agit de clients ayant effectués plus d'un achat (au moins 2), générant ainsi une valeur supérieure à la moyenne des clients (envrion 2 fois plus). Le niveau de satisfaction de ce segment est au niveau de la moyenne de la population (environ 4/5), les clients sont donc satisfait.  
  
**Segment (1)** : les acheteurs **_'Récents à faible valeur'_**.  
Il s'agit de clients ayant commandé une seule fois, plus récemment que la moyenne de la population.  
Les commandes de ces clients sont de montant inférieur relativement aux autres et les acheteurs sont satisfait de ces dernières. Attention néanmoins à la notion de récence, puisque celle ci est relative à la durée sur laquelle est établi l'échantillon !


**Segment (2)** : les acheteurs **_'Mécontents'_**.  
Il s'agit là d'acheteurs ayant un profil moyen (1 seule commande, montant d'environ 180, commande il y a environ 300 jours). Ils se ditinguent uniquement par le fait que la note qu'ils attibuent à leur expérience d'achat est très faible (1,38 en moyenne).  
  
**Segment (3)** : les acheteurs **_'Récents à forte valeur'_**.  
Il s'agit de clients ayant commandé une seule fois, plus récemment que la moyenne de la population.  
Les commandes de ces clients sont de montants relativement supérieurs aux autres et les acheteurs sont satisfait de celles ci. Attention néanmoins à la notion de récence, puisque celle ci est relative à la durée sur laquelle est établie l'échantillon !  
  
**Segment (4)** : les acheteurs **_'Perdus'_**. 
Il s'agit dans ce segment de clients ayant commandé une seule fois, pour une valeur plus faible que la moyenne et ce il y a plus longtemps que la majorité des autres clients. Ce sont néanmoins des clients qui ont été satisfait de leur achat.

```python
segments_products = BDD_CP.merge(right=customers_segments, left_index=True, right_index=True)
display(segments_products.head())
```

On groupe ensuite par segment et l'on exprime les résultats en terme de % du nombre de produits vendus par segment :

```python
segments_products_grouped = segments_products.groupby('segment').sum()
display(segments_products_grouped)
```

```python
segments_products_grouped = segments_products_grouped.div(segments_products_grouped.sum(axis=1),
                                                          axis=0) *100
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(segments_products_grouped)
```

On récupère pour chaque segment les noms des 5 catégories d'objets les plus vendus

```python
top_products_indices = np.argsort(-segments_products_grouped.values, axis=1)[:, :5]
top_products = pd.DataFrame(segments_products_grouped.columns.values[top_products_indices], 
                            columns = ['1st','2nd','3rd', '4th', '5th'])
display(top_products)
```

On récupère pour chaque segment le % d'objets vendus pour les 5 catégories d'objets les plus vendus

```python
top_products_score = pd.DataFrame([segments_products_grouped.iloc[i, top_products_indices[i]].values for i in range(0,5)])
```

```python
fig = plt.figure(figsize=(14,10))
plt.subplots_adjust(hspace=0.8, wspace=0.3)
for i in range(0,5):
    ax = fig.add_subplot(2,3,i+1)
    ax.bar(x=top_products.iloc[i,:], height=top_products_score.iloc[i,:])
    ax.tick_params(axis='x', labelrotation=90)
    ax.set_title('Segment ' + str(i))
    ax.set_ylabel('% of total products sold')
```

Au final on retrouve relativement les mêmes catégories de produits dans les différents segments, et qui correspondent à celles les plus représentées dans le dataset. Les différences en termes de % de produits vendus par rapport au total sont égalements non significatives.

## Actions possibles

**Segment (0)** : les acheteurs **_'récidivistes'_** :    
Pour ce segment, les clients sont déjà relativement fidèles à Olist, si on les compare au reste de la population. Pour encourager ces clients à poursuivre leurs achats sur le store Olist, il pourrait être intéressant de les remercier avec un geste commercial comme un coupon de réduction sur un prochain achat par exemple. Il pourrait également être intéressant de transformer ces clients en 'embassadeurs' du store Olist en mettant en place un système de parrainage.

  
**Segment (1)** : les acheteurs **_'Récents à faible valeur'_** :  
Pour ce segment, si la segmentation a été effectuée de manière suffisament fréquence, la notion de récente du client est la plus intéressante à exploiter. Il serait possible de juste faire parvenir à ces clients un rappel avec une suggestion de produits se basant sur leur dernier achat comme par exemple : 'Ne nous vous avons pas vu depuis un moment, découvrez .....'


**Segment (2)** : les acheteurs **_'Mécontents'_** :  
Pour cette catégorie de clients, il est essentiel de comprendre l'origine de leur mécontentement.
Il existe sans doutes des sous catégories au sein de ce segment qui sont facilement identifiables (livraison en retard, mauvaise qualité produit ...). Ainsi il est éventuellement possible de reprendre contact avec ces utilisateurs en leur annoncant une amélioration de la qualité du service sur le point qui avait fait défaut la première fois.

  
**Segment (3)** : les acheteurs **_'Récents à forte valeur'_** :  
On peut adopter pour cette catégorie la même démarche que pour le segment (1), mais en adaptant les produits poussés sur une valeur plus haute, et en proposant également des solutions de financements. Il s'agit du segment le plus important en termes de chiffre d'affaire et donc le plus intéressant à fidéliser et à faire passer dans le segment (0).
  
**Segment (4)** : les acheteurs **_'Perdus'_** :  
On peut également pour ce segment adopter une démarche similaire à celle du segment (1), mais en leur faisant plutôt parvenir un 'bilan' des nouveautés mises en place depuis qu'ils ne sont pas venus sur le store.
